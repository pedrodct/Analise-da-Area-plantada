```python
# Selecionando a planilha a ser usada

area_plantada_ibge1= pd.read_excel('tabela1612.xlsx', sheet_name='Área plantada')

area_plantada_ibge1.info()
```

```python
# Olhando os dados precisamos organizar o modelo, para isso, vamos ajustar o cabeçalho para entender melhor as colunas e excluir dados desncessarios
area_plantada_ibge1.head(50)
```

```python
# Renomeando o cabecalho e dropando linhas desnecessarias

new_columns = {'Unnamed: 1': 'uf', 
               'Unnamed: 2': '2019',
               'Unnamed: 3': 'Unidade de medida',
               'Unnamed: 4': '2020',
               'Unnamed: 5': 'Unidade de medida',
               'Unnamed: 6': '2021',
               'Unnamed: 7': 'Unidade de medida'}

area_plantada_ibge2 = area_plantada_ibge1.rename(columns=new_columns)
area_plantada_ibge2 = area_plantada_ibge2.drop([0, 1, 2, 3])
area_plantada_ibge2 = area_plantada_ibge2.reset_index(drop=True)
area_plantada_ibge2 = area_plantada_ibge2.drop(index=range(29, len(area_plantada_ibge2)))


area_plantada_ibge2.head(35)
```

```python
# Importando a base de dados da conab

url = "https://portaldeinformacoes.conab.gov.br/downloads/arquivos/SerieHistoricaGraos.txt"

conab = pd.read_csv(url, sep=";", encoding="ISO-8859-1")

display(conab)
```

```python
# filtrando o dataframe para apenas soja

print(conab['produto'].value_counts())
conab.describe()
conab.info()
```

```python
# Tive dificuldade de filtrar o dataframe transformei ele para Excel para buscar uma solução diferente

conab.to_excel('conab_soja.xlsx', index=False)
```

```python
# Importando o dado em formato excel
conab_soja= pd.read_excel('conab_soja.xlsx')

display(conab_soja)
```

```python
# Filtrando linhas que apresentem apenas soja

filtro_produto = conab_soja['produto'].str.contains('SOJA', case=False)
filtro_ano = conab_soja['ano_agricola'] >= '2018/19'
novo_df = conab_soja[filtro_produto & filtro_ano]

display(novo_df)
```

```python
# Criando uma coluna com o valor final de area plantada

novo_df.loc[:, 'area_plantada_ha'] = novo_df['area_plantada_mil_ha'] * 1000

display(novo_df)
```

```python
# Entendendo como estão os formatos do DF

novo_df.info()
```

```python
print(novo_df['ano_agricola'].value_counts())
```

```python
# Tive dificuldades para criar uma coluna com apenas o ano de produção

substituicoes = {'2018/19': 2019, '2019/20': 2020, '2020/21': 2021, '2021/22': 2022, '2022/23': 2023}

novo_df['ano_numerico'] = novo_df['ano_agricola'].apply(lambda x: substituicoes.get(x, None))

display(novo_df)
```

```python
# Como solução usei minha experiencia no excel, e "criei" a função esquerda com 4 digitos

novo_df['ano_agricola_novo'] = novo_df['ano_agricola'].str.slice(stop=4)
display(novo_df)
```

```python
#Adicionei 1 ano a mais para ficar na mesma base do ibge

novo_df['ano_agricola_novo'] = novo_df['ano_agricola_novo'].astype(int) + 1

display(novo_df)
```

```python
# selecionei as linhas com os valores 2022 ou 2023 e dropamos
linhas_para_dropar = novo_df.loc[(novo_df['ano_agricola_novo'] == 2022) | (novo_df['ano_agricola_novo'] == 2023)].index
novo_df.drop(linhas_para_dropar, inplace=True)

display(novo_df)
```

```python
# Criar um novo pivot.table com colunas no cabeçalho sendo UF e mais 3 colunas com unique do ano_agricola_novo sendo o que será prenchindo nestas colunas será a area_plantada_ha em relação a cada ano_agricola_novo e UF

pivot_table = pd.pivot_table(novo_df, index=['uf'],columns=['ano_agricola_novo'], values=['area_plantada_ha'], aggfunc=np.sum)


brasil_somas = pivot_table.sum(axis=0)
pivot_table = pivot_table.append(pd.Series(name="Brasil", data=brasil_somas).to_frame().T)

display(pivot_table)

```

```python
# Trocando as siglas dos estados.
uf_dic = {'AC': 'Acre', 'AL': 'Alagoas', 'AP': 'Amapá', 'AM': 'Amazonas', 'BA': 'Bahia', 'CE': 'Ceará', 'DF': 'Distrito Federal', 'ES': 'Espírito Santo', 'GO': 'Goiás', 'MA': 'Maranhão', 'MT': 'Mato Grosso', 'MS': 'Mato Grosso do Sul', 'MG': 'Minas Gerais', 'PA': 'Pará', 'PB': 'Paraíba', 'PR': 'Paraná', 'PE': 'Pernambuco', 'PI': 'Piauí', 'RR': 'Roraima', 'RO': 'Rondônia', 'RJ': 'Rio de Janeiro', 'RN': 'Rio Grande do Norte', 'RS': 'Rio Grande do Sul', 'SC': 'Santa Catarina', 'SP': 'São Paulo', 'SE': 'Sergipe', 'TO': 'Tocantins'}
pivot_table = pivot_table.rename(index=uf_dic)

display(pivot_table)
```

```python
# Vamos fazer o merger das duas planilhas para comparar e fazer a analise.

pivot_table.head(5)
```

```python

area_plantada_ibge2.info()
```

```python
# juntando as duas df
df_final = pd.concat([pivot_table, area_plantada_ibge2], axis=1)

display(df_final)
```

```python
df_final1 = df_final.sort_values(by=( 'area_plantada_ha', 2021), ascending=False, na_position='last')

display(df_final1)
```
